```python
import pandas as pd
import numpy as np



# lets create 1000 random numbers
age = np.random.randint(18, 60, size=1000)
len(age)
```

```python
# lets compute population mean
pop_mean = np.mean(age)
pop_mean
```

```python
# lets create sample of size 50
sample = np.random.choice(age, size=50, replace=False)
print(len(sample))
sample_mean = np.mean(sample)
sample_mean
```

```python
# one more sample
sample2 = np.random.choice(age, size=50, replace=False)
sample2_mean = np.mean(sample2)
sample2_mean
```

```python
#one mopre sample
sample3 = np.random.choice(age, size=50, replace=False)
sample3_mean = np.mean(sample3)
sample3_mean
```

```python
total_samples_mean = (sample_mean + sample2_mean + sample3_mean) / 3
total_samples_mean

# Conclusion: As we take more samples, the sample mean approaches the population mean.
```

```python
age = np.random.normal(loc=50, scale=10, size=1000)
len(age)
pop_mean = np.mean(age)
pop_mean
```

```python
# plot it using matplotlib along with kde

import matplotlib.pyplot as plt 
import seaborn as sns

sns.histplot(age, bins=30, kde=True)
plt.axvline(pop_mean, color='red', linestyle='dashed', linewidth=2)
plt.title('Age Distribution with Population Mean')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 68-95-99.7 rule states that for a normal distribution:
# Approximately 68% of the data falls within one standard deviation () of the mean.
# Approximately 95% of the data falls within two standard deviations (2) of the mean.
# Approximately 99.7% of the data falls within three standard deviations (3) of the mean.

# example
mean = 50
std_dev = 10
one_std_lower = mean - std_dev
one_std_upper = mean + std_dev
two_std_lower = mean - 2 * std_dev
two_std_upper = mean + 2 * std_dev
three_std_lower = mean - 3 * std_dev
three_std_upper = mean + 3 * std_dev    
print(f"68% of data falls between {one_std_lower} and {one_std_upper}")
print(f"95% of data falls between {two_std_lower} and {two_std_upper}")
print(f"99.7% of data falls between {three_std_lower} and {three_std_upper}")
```

```python
# lets take thsirt example
# Tshirt small, medium, large sizes
# 5000 - how many small, medium, large tshirts to produce based on normal distribution of chest sizes
chest_sizes = np.random.normal(loc=38, scale=4, size=5000)  # mean chest size 38 inches, std dev 4 inches


# % of people with chest size less than 34 inches (small size)
small_size_threshold = 34
percent_small = (np.sum(chest_sizes < small_size_threshold) / len(chest_sizes)) * 100
print(f"Percentage of people with chest size less than {small_size_threshold} inches: {percent_small:.2f}%")
```

```python
# plot it with size categories
sns.histplot(chest_sizes, bins=30, kde=True)
plt.axvline(small_size_threshold, color='blue', linestyle='dashed', linewidth=2, label='Small Size Threshold (34 inches)')
plt.title('Chest Size Distribution with Small Size Threshold')
plt.xlabel('Chest Size (inches)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# number of small, medium, large tshirts to produce
num_small = int((percent_small / 100) * 5000)
num_medium = int(0.5 * 5000)  # assuming 50% medium size
num_large = 5000 - num_small - num_medium
num_small, num_medium, num_large
```

```python
# lets talk about skewness and kurtosis
# Skewness measures the asymmetry of the probability distribution of a real-valued random variable about its mean.

# eg from bank loan data
loan_amounts = np.random.normal(loc=5000, scale=1500, size=1000)
# introduce some skewness
loan_amounts = np.concatenate([loan_amounts, np.random.normal(loc=15000, scale=2000, size=100)])  # adding some high loan amounts
# plot it
sns.histplot(loan_amounts, bins=30, kde=True)
plt.title('Loan Amount Distribution with Skewness')
plt.xlabel('Loan Amount ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
# left skewd data
left_skewed_data = np.random.exponential(scale=2, size=1000)
# plot it
sns.histplot(left_skewed_data, bins=30, kde=True)
plt.title('Left Skewed Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# real example and conclusion
# e.g., income distribution in a population is often right-skewed, with a small number of individuals earning significantly more than the majority.
# In contrast, the distribution of heights in a population is typically symmetric and follows a normal distribution
```

```python
income = np.random.exponential(scale=50000, size=1000)
# plot it along with mean
    
sns.histplot(income, bins=30, kde=True)
plt.axvline(income.mean(), color='r', linestyle='dashed', linewidth=1)
plt.title('Income Distribution with Right Skewness')
plt.xlabel('Income ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
# left skewed height data
height = np.random.normal(loc=65, scale=3, size=1000)
# introduce left skewness
height = np.concatenate([height, np.random.normal(loc=55, scale=2, size=100)])  # adding some low height values
# plot it
sns.histplot(height, bins=30, kde=True)
plt.title('Height Distribution with Left Skewness')
plt.xlabel('Height (inches)')
plt.ylabel('Frequency')
plt.show()
```

```python
stock_prices = np.random.normal(loc=100, scale=20, size=1000)
# introduce some skewness
stock_prices = np.concatenate([stock_prices, np.random.normal(loc=200, scale=30, size=100)])  # adding some high stock prices
# plot it
sns.histplot(stock_prices, bins=30, kde=True)
plt.title('Stock Price Distribution with Skewness')
plt.xlabel('Stock Price ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
# what is kurtosis?
# Kurtosis measures the "tailedness" of the probability distribution of a real-valued random variable.
# High kurtosis indicates heavy tails and a sharp peak, while low kurtosis indicates light tails and a flatter peak.
# Example: A normal distribution has a kurtosis of 3. Distributions with kurtosis greater
# than 3 are called leptokurtic (heavy-tailed), while those with kurtosis less than 3 are called platykurtic (light-tailed).
```
